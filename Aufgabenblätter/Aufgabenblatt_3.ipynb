Import aller wichtigen Bibliotheken:

```python
import re
import os
import nltk
from nltk.corpus import PlaintextCorpusReader
```

Erstellung eines Korpus:

```python
corpus_dir = os.getcwd() + "/data"
corpus_txts = ".+\.txt"
korpus = PlaintextCorpusReader(corpus_dir, corpus_txts, encoding="utf-8")
print(korpus.fileids())
```

# Aufgabe 1
- Definieren Sie eine Funktion `voc_con`, die in einer Zeichenkette das (numerische) Verhältnis von Konsonanten zu Vokalen berechnet und geben Sie die Werte für beide Texte an.

1. Die Funktion nimmt einen Parameter `text` der einem Dateinamen in unserem Korpus entsprechen muss.
2. Dieser `text` wird in einen `str-Type` umgewandelt.
3. Es wird eine Liste mit allen Vokalen in `voc` abgelegt.
4. Es wird eine Liste mit allen Konsonanten in `con` abgelegt.
5. Es wird das Verhältnis der Längen der Listen `voc` und `con` zurückgegeben.

```python
def voc_con(text):
    text = korpus.raw(text)
    voc = re.findall("[AEIOUaeiouÄÖÜäöü]", text)
    con = re.findall("[^-AEIOUaeiouÄÖÜäöü,\s\.;_:*?!\"]", text)
    return len(voc) / len(con)
```

```python
voc_con("trump.txt")
```

```python
voc_con("corona.txt")
```

# Aufgabe 2
- Definieren Sie die Funktion `vc_pattern`, die eine Zeichenkette als Argument nimmt und jeden Vokal (Konsonant) durch ein ’V’ (’C’) ersetzt. Zeichen, die weder einen Vokal noch einen Konsonanten repräsentieren, werden entfernt.


1. Die Funktion nimmt einen Parameter `text` der einem Dateinamen in unserem Korpus entsprechen muss.
2. Der `text` wird in einen String umgewandelt.
3. Es werden die Sonderzeichen entfernt.
4. Es werden alle Konsonanten in ein `C` umgewandelt.
5. Es werden alle Vokale in ein `V` umgewandelt.
6. Die Funktion gibt den bearbeiteten Text als Wert zurück.

```python
def vc_pattern(text):
    text = korpus.raw(text)
    text = re.sub("[-,\.;_:*?!\"\d\n]", "", text)
    text = re.sub("[^AEIOUaeiouÄÖÜäöü\s]", "C", text)
    text = re.sub("[AEIOUaeiouÄÖÜäöü]", "V", text)
    return text
```

```python
vc_pattern("trump.txt")
```

```python
vc_pattern("corona.txt")
```

# Aufgabe 3
Definieren Sie die Funktionen satzlänge und wortlänge, die die durchschnittliche Satz- und Wortlänge der Sätze/Wörter in einem Text oder Korpus berechnet. Die Länge eines Satzes können Sie hier mit der Zahl seiner Token identifizieren; die Länge eines Wortes als Zahl seiner Grapheme.
Welche Werte erhalten Sie für die beiden Texte?

```python
def satzlaenge(text):
    return round(len(korpus.words(text)) / len(korpus.sents(text)))
```

```python
satzlaenge("trump.txt")
```

```python
satzlaenge("corona.txt")
```

```python
def wortlaenge(text):
    return round(len(korpus.raw(text)) / len(korpus.words(text)))
```

```python
wortlaenge("trump.txt")
```

```python
wortlaenge("corona.txt")
```
